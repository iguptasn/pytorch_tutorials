```python
import torch
from torch import nn
```

```python
# Need Neural network definition for loading models from model weights or model weights plus structures

class NeuralNetwork(nn.Module):
    def __init__(self):
        super().__init__()
        self.flatten = nn.Flatten()
        self.linear_relu_stack = nn.Sequential(
            nn.Linear(28*28, 512),
            nn.ReLU(),
            nn.Linear(512,512),
            nn.ReLU(),
            nn.Linear(512,10),
        )
    def forward(self, x):
        x = self.flatten(x)
        logits = self.linear_relu_stack(x)
        return logits
```

```python
model = NeuralNetwork()
model
```

```python
# loading models with model weights

model_weights = torch.load('./models/tutorial6_model_weights.pth')
model.load_state_dict(model_weights)
model.eval(), list(model.parameters())
```
